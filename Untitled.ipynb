```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Stock_dataa.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.info()
df.columns = df.columns.str.lower()
df.date = pd.to_datetime(df.date, errors='raise', infer_datetime_format=True)
df.date
```

```python
train = df[df.date < '2015-01-01']
test = df[df.date > '2014-12-31']
```

```python
#some data preprocessing
#take only alphabets from the sentences
data = df.iloc[:, 2:26]
# data
data.replace(to_replace='[^a-zA-Z]', regex=True, value=' ', inplace=True)
```

```python
#make all texts smaller
for col in data.columns:
#     print(type(data[col]))
    data[col] = data[col].str.lower()
```

```python
data
```

```python
# convert each data point i.e. combine headings 
# from all the newspaper for single index
headlines = []
for row in range(df.shape[0]):
    headlines.append([' '.join(str(x) for x in data.iloc[row, :25])])
```

```python
headlines[0]
```

```python
# for x in data.iloc[0, :25]:
#     print(str(x))
```

```python
#using bag of words and random forest classifier
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
#headlines is two dimentioanl convert it into 1D array
headlines = np.array(headlines).flatten()
headlines.shape
```

```python
vectorizer = CountVectorizer(ngram_range=(2, 2))
trained_vectorizer = vectorizer.fit_transform(headlines[:3723])
# vectorizer.fit_transform(headlines)
```

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(trained_vectorizer, train['label'])
```

```python
#make prediction
from sklearn.metrics import accuracy_score
y_pred = rfc.predict(vectorizer.transform(headlines[3723:]))
accuracy_score(test['label'], y_pred)
```
